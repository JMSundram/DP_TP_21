```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('science') # pip install git+https://github.com/garrettj403/SciencePlots.git
```

```python
# Settings
dpi = 400
save = True
```

```python
# Load data
df = pd.read_csv('INDKP105.csv', sep=';', header=None, encoding='latin-1')
df = df.drop(df.columns[0], axis=1)
df.columns = ['alder', 'indk']
df['alder'] = df['alder'].str.replace(' Ã¥r', '')
df['indk'] = df['indk']/1000
```

```python
# Create figure
plt.figure(figsize=(4,3),dpi=dpi)
plt.plot(df['alder'][:-2],df['indk'][:-2])
plt.grid(alpha=1/3) # Make grid
plt.minorticks_off() # Remove minor ticks
plt.tick_params(top=False, right=False) # Remove major ticks on top and right 
plt.xlabel('Age (years)')
plt.ylabel('Income (1000 DKK)')
if save: plt.savefig(r'fig_lifecycle.png')
```
