```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('science') # pip install git+https://github.com/garrettj403/SciencePlots.git
```

```python
# Settings
dpi = 400
save = True
```

```python
# Data
# Source for number of early payments by age group: "Faktaark udbetaling af indefrosne feriemidler"
# Source for number of wage earners in each age group: Table RAS201 at Statistikbanken.dk
ages = ['-19','20-24','25-29','30-34','35-39','40-44','45-49','50-54','55-59','60-']
num_early = [107_668, 232_508, 269_424, 260_005, 231_211, 244_893, 262_662, 254_968, 229_937, 209_456]
num_total = [44_691+147_026, 248_832, 287_567, 279_270, 268_379, 305_473, 328_338, 338_720, 306_356, 225_343 + 41_261 + 102_866]
frac_early = np.array(num_early)/np.array(num_total)
```

```python
# Graph
plt.figure(figsize=(5,3),dpi=dpi)
plt.bar(ages[1:],frac_early[1:],width=0.5,zorder=3)
plt.grid(alpha=1/3)
plt.minorticks_off() 
ax = plt.gca()
ax.tick_params(top=False, right=False)
ax.grid(zorder=0)
plt.xlabel('Age (years)')
plt.ylabel('Fraction of early payments')
if save: plt.savefig(r'fig_frac_early_vs_age')
```
