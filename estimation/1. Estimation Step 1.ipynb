```python
import numpy as np
import time
from model import run_model
from params import par_class
```

```python
# Data
THs = [27-25, 42-25, 57-25]
zs  = [0.937, 0.802, 0.751]
```

```python
# Compute objective function over coarse grid
t0 = time.time()
rhos = np.linspace(0.4, 8.0, 20)
betas = np.concatenate([[0, 0.75], np.linspace(0.76, 1.0, 25)])
msds_br = np.nan + np.zeros((len(betas),len(rhos)))
for i,beta in enumerate(betas):
    for j,rho in enumerate(rhos):
        par = par_class()
        par.Nm = 80
        par.Nh = 12
        par.simN = 10_000
        par.beta = beta
        par.rho = rho
        zs_sim = []
        for TH in THs:
            par.TH = TH
            _, sim = run_model(par)
            z_mean = sim.z.mean()
            zs_sim.append(z_mean)
        mean_squared_dist = np.mean((np.array(zs) - np.array(zs_sim))**2)
        msds_br[i,j] = mean_squared_dist
        print(f'beta = {beta:.3f}, rho={rho:.3f}, and objective = {mean_squared_dist:.5f}')
t1 = time.time()
print(f'Time = {t1-t0:.3f}')
```

```python
# Save objective function as csv
np.savetxt('rhos.csv', rhos, delimiter=',')
np.savetxt('betas.csv', betas, delimiter=',')
np.savetxt('msds_br.csv', msds_br, delimiter=',')
```
